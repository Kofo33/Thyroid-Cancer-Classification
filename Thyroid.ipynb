Encoding the data using Label Encoding

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output=False, drop='first')  # drop='first' for efficiency
```

```python
# Define categorical columns
categorical_cols = [
    'Gender', 'Country', 'Ethnicity', 'Family_History',
    'Radiation_Exposure', 'Iodine_Deficiency', 'Smoking',
    'Obesity', 'Diabetes', 'Thyroid_Cancer_Risk', 'Diagnosis'
]
```

```python
# Apply OneHotEncoder
encoded_data = ohe.fit_transform(df[categorical_cols])
```

```python
# Convert to DataFrame
encoded_df = pd.DataFrame(encoded_data, columns=ohe.get_feature_names_out(categorical_cols))
```

```python
# Drop original categorical columns
df = df.drop(columns=categorical_cols)
```

```python
# Concatenate with the original DataFrame
df = pd.concat([df, encoded_df], axis=1)
```

Using Train-Test Split, training the model using Logistic Regression

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop('Diagnosis_Malignant', axis = 1)
y = df['Diagnosis_Malignant']
```

Splitting the data into train and test, 70% and 30% respectively.

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.preprocessing import StandardScaler

# Assuming x_train (212691, 17), y_train (212691,), x_test, y_test
# Check the shape of x_train
print("Shape of x_train:", x_train.shape)  # Add this line for debugging

# If x_train.shape[1] is not 17, you need to adjust the input_shape
input_shape = x_train.shape[1]  # Get the actual number of features

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

model = Sequential()
# Update input_shape to match the actual number of features
model.add(Dense(128, input_shape=(input_shape,), activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01)))
model.add(BatchNormalization())
model.add(Dropout(0.3))
model.add(Dense(64, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.3))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# ... (Rest of the code remains the same)
```

```python
model.summary()
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
              loss='binary_crossentropy',
              metrics=['accuracy', tf.keras.metrics.AUC()])

early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
history = model.fit(x_train_scaled, y_train,
                    epochs=100,
                    batch_size=64,
                    validation_split=0.2,
                    callbacks=[early_stopping])

test_loss, test_acc, test_auc = model.evaluate(x_test_scaled, y_test)
print(f"Test Accuracy: {test_acc}, Test AUC: {test_auc}")

# Plot losses
import matplotlib.pyplot as plt
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.legend()
plt.show()
```
